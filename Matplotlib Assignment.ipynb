Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

**Matplotlib** is a widely used data visualization library in Python. It provides a way to create a variety of high-quality static, interactive, and animated visualizations in Python. Matplotlib is particularly useful for creating graphs, charts, and plots to represent data in a visually appealing and informative manner.

Matplotlib's `pyplot` module is a collection of functions that provide a simple interface for creating various types of plots. Here are five types of plots that can be created using the `pyplot` module of Matplotlib:

1. **Line Plot**: A line plot is used to visualize data points connected by straight lines. It's commonly used to display trends over time or to show the relationship between two variables.

2. **Scatter Plot**: A scatter plot is used to visualize the distribution of data points and the relationship between two variables. Each data point is represented as a dot on the graph.

3. **Bar Plot**: A bar plot, also known as a bar chart, is used to compare categorical data using rectangular bars. It's often used to show comparisons between different categories or to display frequencies.

4. **Histogram**: A histogram is used to display the distribution of a continuous variable by dividing the data into bins and showing the frequency of data points in each bin.

5. **Pie Chart**: A pie chart is a circular chart that is divided into slices to represent the proportion of different categories within a whole. It's commonly used to show the composition of a whole in terms of percentages.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot. <br>
import numpy as np <br>
np.random.seed(3) <br>
x = 3 + np.random.normal(0, 2, 50) <br>
y = 3 + np.random.normal(0, 2, len(x)) <br>
Note: Also add title, xlabel, and ylabel to the plot.

Scatter Plot: A scatter plot is used to visualize the distribution of data points and the relationship between two variables. Each data point is represented as a dot on the graph.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('X Axis')
plt.xlabel('Y Axis')
plt.xlabel('Scatter Graph')
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. <br>
Use the following data:<br>
import numpy as np<br>
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])<br>
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])<br>
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])<br>
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])

# Data for line plots
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line Plot 4')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot. <br>
box1 = np.random.normal(100, 10, 200) <br>
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation that displays the summary of a set of data in terms of its distribution, spread, and potential outliers. It provides a compact way to visualize the distribution of a dataset by showing the median, quartiles, and potential outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.boxplot(data, labels=labels)
plt.title('Box Plot')
plt.ylabel('Value')
plt.xlabel('Box')
plt.grid(True)
plt.show()
```
