Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. **Scatter Plot:**
   - Use: Scatter plots display individual data points as dots. They are used to visualize the relationship between two numerical variables, showing how one variable changes in relation to the other.
   
2. **Line Plot:**
   - Use: Line plots display data points connected by lines. They are used to show trends or patterns over a continuous interval, such as time. Line plots are commonly used for time series data.

3. **Bar Plot:**
   - Use: Bar plots display categorical data with rectangular bars. They are used to compare values across different categories. Bar plots can be used to show counts, frequencies, or aggregated values of a categorical variable.

4. **Histogram:**
   - Use: Histograms display the distribution of a single numerical variable. They group data into bins and show the frequency or count of data points in each bin. Histograms help in understanding the data's underlying distribution.

5. **Box Plot (Box-and-Whisker Plot):**
   - Use: Box plots display the summary statistics of a numerical variable's distribution. They show the median, quartiles, and potential outliers in the data. Box plots are useful for identifying the spread and skewness of data, as well as detecting outliers.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sb
```

```python
fmri = sb.load_dataset('fmri')
```

```python
fmri
```

```python
sb.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='region')
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights_data = sns.load_dataset("flights")
flights_pivot = flights_data.pivot_table(index="month", columns="year", values="passengers")
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Flights Passengers Heatmap")
plt.show()

```
